```python
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Data: Monitors cost by size in inches
size = np.array([23.8,27.0,28.0,24.0,23.5,34.0,21.5,23.8,24.0,27.0,15.6,32.0,34.1,23.0]).reshape((-1,1))
cost = np.array([139,219,369,141,153,469,147,209,159,199,169,439,439,129])
```

```python
# Scatterplot der Trainingsdaten
plt.scatter(size, cost)
plt.xlabel("size in inches")
plt.ylabel("Price in Euro")
plt.title("Price of Monitors relative to size");
```

```python
# regression
regr = linear_model.LinearRegression()
regr.fit(size, cost);
```

```python
# testdata
testdata_size = np.array([27.0,28.0,24.0,28.0,28.0,23.8,24.0,25.0,28.0,23.8,32.0,34.1,34.0,23.8,21.0]).reshape((-1,1))
testdata_cost = np.array([200,230,189,249,225,150,199,210,400,189,310,390,439,200,199])
```

```python
# prediction
prediction = regr.predict(testdata_size)
```

```python
plt.scatter(testdata_size, testdata_cost,  color='green')
plt.plot(testdata_size, prediction, linewidth=3)
plt.xlabel("size in inches")
plt.ylabel("Price in Euro")
plt.title("Price of Monitors relative to size");
```
