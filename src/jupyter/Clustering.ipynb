```python
import pandas as pd
import matplotlib.pyplot as plot
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
```

```python
data_frame = pd.read_csv('edlich-kmeans-A0.csv')
data_frame.head(10)
```

```python
# cluster center distances
sum_distances = []
K = range(1,10)
for k in K:
    k_means = KMeans(n_clusters=k)
    k_means = k_means.fit(data_frame)
    sum_distances.append(k_means.inertia_)
```

```python
#elbow
plot.title('Elbow for KMeans clustering')
plot.plot(K, sum_distances, 'or-')
plot.xlabel('K')
plot.ylabel('Sum of distances')
plot.show()
```
